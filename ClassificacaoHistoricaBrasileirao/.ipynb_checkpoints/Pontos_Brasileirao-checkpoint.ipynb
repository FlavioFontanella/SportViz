```python
df_br = pd.read_csv('jogosbra3.csv',sep=";",encoding='Latin-1')
season = [x[0:4] for x in list(df_br['date'])]
for i in range(len(season)):
    if season[i]=='2021' and i<8116:
        season[i]='2020'
df_br['season']=season
df_br = df_br[df_br.season>'2002']
df_br = df_br[df_br.tourn!='Copa do Brasil']
a = list(df_br.t1)
for i in range(len(a)):
    if a[i]=='Atletico-MG':
        a[i] = 'Atlético-MG'
    elif a[i]=='Athlético-PR':
        a[i] = 'Athletico-PR'
    elif a[i]=='Avai':
        a[i] = 'Avaí'
    elif a[i]=='Botafogo RJ':
        a[i] = 'Botafogo'
    elif a[i]=='Botafogo-RJ':
        a[i] = 'Botafogo'
    elif a[i]=='Ceara':
        a[i] = 'Ceará'
    elif a[i]=='Chapecoense-SC':
        a[i] = 'Chapecoense'
    elif a[i]=='Criciuma':
        a[i]='Criciúma'
    elif a[i]=='Flamengo RJ':
        a[i] = 'Flamengo'
    elif a[i]=='Goias':
        a[i]='Goiás'
    elif a[i]=='Gremio':
        a[i] = 'Grêmio'
    elif a[i]=='Parana':
        a[i] = 'Paraná'
    elif a[i]=='Paysandu PA':
        a[i] = 'Paysandu'
    elif a[i]=='Sao Caetano':
        a[i]='São Caetano'
    elif a[i]=='Sao Paulo':
        a[i] = 'São Paulo'
    elif a[i]=='Sport Recife':
        a[i] = 'Sport'
    elif a[i]=='Vitoria':
        a[i] = 'Vitória'
    elif a[i]=='America MG':
        a[i] = 'América-MG'
    elif a[i]=='America RN':
        a[i] = 'América-RN'
    elif a[i]=='Atletico GO':
        a[i] = 'Atlético-GO'
    elif a[i]=='Betim':
        a[i] = 'Ipatinga'
df_br.t1 = a

a = list(df_br.t2)
for i in range(len(a)):
    if a[i]=='Atletico-MG':
        a[i] = 'Atlético-MG'    
    elif a[i]=='Athlético-PR':
        a[i] = 'Athletico-PR'
    elif a[i]=='Avai':
        a[i] = 'Avaí'
    elif a[i]=='Botafogo RJ':
        a[i] = 'Botafogo'
    elif a[i]=='Botafogo-RJ':
        a[i] = 'Botafogo'
    elif a[i]=='Ceara':
        a[i] = 'Ceará'
    elif a[i]=='Chapecoense-SC':
        a[i] = 'Chapecoense'
    elif a[i]=='Criciuma':
        a[i]='Criciúma'
    elif a[i]=='Flamengo RJ':
        a[i] = 'Flamengo'
    elif a[i]=='Goias':
        a[i]='Goiás'
    elif a[i]=='Gremio':
        a[i] = 'Grêmio'
    elif a[i]=='Parana':
        a[i] = 'Paraná'
    elif a[i]=='Paysandu PA':
        a[i] = 'Paysandu'
    elif a[i]=='Sao Caetano':
        a[i]='São Caetano'
    elif a[i]=='Sao Paulo':
        a[i] = 'São Paulo'
    elif a[i]=='Sport Recife':
        a[i] = 'Sport'
    elif a[i]=='Vitoria':
        a[i] = 'Vitória'
    elif a[i]=='America MG':
        a[i] = 'América-MG'
    elif a[i]=='America RN':
        a[i] = 'América-RN'
    elif a[i]=='Atletico GO':
        a[i] = 'Atlético-GO'
    elif a[i]=='Betim':
        a[i] = 'Ipatinga'
df_br.t2 = a

#PUNIÇÕES
#2003.04.15 Ponte Preta-1 Internacional+2
#2003.04.29 Ponte Preta-3 Juventude+3
#2003.09.23 Paysandu-4 Corinthians+2 Ponte Preta+3
#2003.10.08 Paysandu-4 São Caetano+3 Fluminense+2
#2004.12.07 São Caetano-24
#2005.05.05 Brasiliense-1 Vasco+2
#2010.08.19 Barueri-3
#2013.12.16 Flamengo-4 Portuguesa-4
#2016.12.11 Chapecoense-1 Atlético-MG-1
#2017.06.01 Santa Cruz-3
#2019.04.24 Sport-1
punish_dates = ['2003.04.15','2003.04.29','2003.09.23','2003.10.08','2004.12.07','2005.05.05','2010.08.19','2013.12.16','2016.12.11','2017.06.01','2019.04.24']
punishes = [[['Ponte Preta',-1],['Internacional',2]],
            [['Ponte Preta',-3],['Juventude',3]],
            [['Paysandu',-4],['Corinthians',2],['Ponte Preta',3]],
            [['Paysandu',-4],['São Caetano',3],['Fluminense',2]],
            [['São Caetano',-24]],
            [['Brasiliense',-1],['Vasco',2]],
            [['Barueri',-3]],
            [['Flamengo',-4],['Portuguesa',-4]],
            [['Chapecoense',-1],['Atlético-MG',-1]],
            [['Santa Cruz',-3]],
            [['Sport',-3]
           ]

teams = list(df_br.t1.drop_duplicates())
teams.sort()

dates1 = list(df_br.date.drop_duplicates())+['2003.04.15','2003.04.29','2003.09.23','2004.12.07','2005.05.05','2010.08.19','2013.12.16','2017.06.01','2019.04.24']
dates1.sort()
dates = ['2003.03.28']
for i in range(len(dates1)):
    dates.append(dates1[i])
    
pts_array=[0]*len(dates)*len(teams)
goals_array=[0]*len(dates)*len(teams)
matches_array=[0]*len(teams)
dt_count = 0

for date in dates:
    for i in range(len(df_br[df_br.date==date])):
        t1 = df_br[df_br.date==date].iat[i,2]
        g1 = df_br[df_br.date==date].iat[i,3]
        g2 = df_br[df_br.date==date].iat[i,4]
        t2 = df_br[df_br.date==date].iat[i,5]
        #print(date,teams.index(t1),t1,g1,'x',g2,t2,teams.index(t2))
        matches_array[teams.index(t1)]+=1
        matches_array[teams.index(t2)]+=1
        goals_array[teams.index(t1)*len(dates)+dt_count]=g1
        goals_array[teams.index(t2)*len(dates)+dt_count]=g2
        if g1>g2: 
            pts_array[teams.index(t1)*len(dates)+dt_count]=3
        elif g1==g2:
            pts_array[teams.index(t1)*len(dates)+dt_count]=1
            pts_array[teams.index(t2)*len(dates)+dt_count]=1
        else:
            pts_array[teams.index(t2)*len(dates)+dt_count]=3
    dt_count+=1
    
for i in range(len(punish_dates)):
    for punish in punishes[i]:
        ajustar = teams.index(punish[0])*len(dates)+dates.index(punish_dates[i])
        pts_array[ajustar]+=punish[1]

pts_matrix = []
goals_matrix = []

for i in range(len(teams)):
    arr = [0]*len(dates)
    arr2 = [0]*len(dates)
    arr[0]=pts_array[len(dates)*i]
    arr2[0]=goals_array[len(dates)*i]
    for j in range(len(dates)-1):
        arr[j+1]=arr[j]+pts_array[len(dates)*i+j+1]
        arr2[j+1]=arr2[j]+goals_array[len(dates)*i+j+1]
    pts_matrix.append(arr)
    goals_matrix.append(arr2)
    
dates1 = [dates[i].replace('.','-') for i in range(len(dates))]
dates2 = []

for str_date in dates1:
    date = datetime.strptime(str_date, '%Y-%m-%d').date()
    dates2.append(date)

df1 = pd.DataFrame(index=dates2)
df2 = pd.DataFrame(index=dates2)

for i in range(len(teams)):
    df1[teams[i]]=pts_matrix[i]
    df2[teams[i]]=goals_matrix[i]

colorteams = ["#81f040"]+["#e04040"]+["#800000"]+["#e03030"]+ ["#505050"]+["#70c0e0"]+["#4081d0"]+["#000000"]+["#808080"]+["#000000"]+["#000000"]+["#000000"]+["#000000"]+["#000000"]+ ["#e0e0e0"]+["#40e080"]+["#d0c010"]+["#0020c0"]+["#c3f020"]+["#b0c0b0"]+ ["#c3281e"]+["#80c040"]+["#6020f0"]+["#20a000"]+["#30c1ff"]+["#008040"]+ ["#ff3030"]+["#30ff60"]+["#109020"]+["#5040f0"]+["#40b1d0"]+ ["#202020"]+["#ffffff"]+["#a01000"]+["#4000a0"]+["#e0acac"]+["000000"]+["#b02000"]+ ["yellow"]+["pink"]+["orange"]+["brown"]

'''     América-MG', 'América-RN', 'Athletico-PR', 'Atlético-GO',
       'Atlético-MG', 'Avaí', 'Bahia', 'Barueri', 'Botafogo',
       'Bragantino', 'Brasiliense', 'CSA', 'Ceará', 'Chapecoense',
       'Corinthians', 'Coritiba', 'Criciúma', 'Cruzeiro', 'Cuiabá', 'Figueirense',
       'Flamengo', 'Fluminense', 'Fortaleza', 'Goiás', 'Grêmio', 'Guarani',
       'Internacional','Juventude', 'Palmeiras', 'Paraná', 'Paysandu', 
       'Ponte Preta', 'Santos', 'Sport', 'São Caetano','São Paulo', 'Vasco', 'Vitória'
       '''
```
